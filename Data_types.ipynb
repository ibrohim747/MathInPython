```python
import numpy as np
```

```python
alist = [0, 1, 2, 3, 4, 5]
slist = ['0','1','2','3','4','5']

print(alist)
print(slist)
```

```python
alist + slist
```

```python
alist_np = np.array(alist)

print(alist)
print(alist_np)
```

```python
alist * 3
```

```python
alist_np * 3
```

```python
alist + 3   #Error
alist + [3]   #Correct
#List can only + to list
```

```python
listlist = [
    3,
    ['3','4','5'],
    5.5,
    [4,5,6]
    ]
listlist
```

```python
b = 0
for i in listlist:
    el_type = str(type(i))[8:-2]
    print("List element %g is %r and is of type %r" %(b, i, el_type))
    b += 1
```
