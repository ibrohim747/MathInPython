```python
display(Math(r'a^2 + b^2 = c^2'))
print("")

display(Math(r'x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}'))
print("")

display(Math(r'\frac{d}{dx} \left( x^n \right) = nx^{n-1}'))
print("")

display(Math(r'\int_a^b f(x) \, dx'))
print("")

display(Math(r'f(x) = \sum_{n=0}^{\infty} \frac{f^{(n)}(a)}{n!} (x-a)^n'))
```

```python
display(Math('3^2 = 9'))
display(Math('3^2 \\times 3^4 = 3^{2+4}'))
```

```python
x = 5
y = 5.1

#1 x**3/4 * 4**y
display(Math('x^3/4 \\times 4^y = %g' %(x**(3/4) * 4**y)))
print("")

#2 (3**3)/(x**y)
display(Math('\\frac{3^3}{x^y} = %g' %((3**3)/(x**y))))
print("")

#3 10**x-4
display(Math('10^{x-4} = %g' %(10**(x-4))))
```
