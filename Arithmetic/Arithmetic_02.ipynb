# For-loops

```python
for i in (1, 2, 3, 4, 5):
    print(i)
```

```python
for i in (1, 2, 3, 4, 5):
    print(i**2)
```

```python
for querty in range(0,3):
  print('This is the %gth iteration.' %querty)
```

```python
for i in range(0,10):
  print('2 to the power of %g is %g' %(i, 2**i))
```

```python
res = 4 * 5 / (7 + 3) #2
res_2 = 9 / (3 + 6) - 1 #0
print(round(res, 0), res_2)
```
