```python
import sympy as sym
from IPython.display import display, Math
```

```python
mu, alpha, sigma = sym.symbols('mu, alpha, sigma')

expr = sym.exp( (mu-alpha)**2 / (2*sigma**2) )
display(expr)
```

```python
hello = sym.symbols('hello')
display(hello/3)
```

```python
x, y = sym.symbols('x, y')
expr = x + 4 + 2 * y
expr.subs(x, 6) #x = 6
#expr.subs({x: 6, y:3}) #x = 6, y = 3
```

```python
expr = 3 / x
sym.latex(expr) #LaTeX code
```

```python
display(Math(sym.latex(expr)))
```

```python
x, y, z = sym.symbols('x, y, z')
ex = x**y * x**z
display(ex)
display(sym.simplify(ex))
```

```python
exp1 = x**y * x**z
exp2 = x**y / x**z
exp3 = x**y * y**z

display(Math('%s = %s' %(sym.latex(exp1), sym.latex(sym.simplify(exp1)))))
display(Math('%s = %s' %(sym.latex(exp2), sym.latex(sym.simplify(exp2)))))
display(Math('%s = %s' %(sym.latex(exp3), sym.latex(sym.simplify(exp3)))))
```

```python
#Equal
sym.Eq(4, 2+2) #True
```

```python
sym.Eq(exp1, sym.simplify(exp1))
```

```python
sym.Eq(exp1 - sym.simplify(exp1), 0)
```
