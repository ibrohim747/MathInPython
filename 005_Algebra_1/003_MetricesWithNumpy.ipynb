```python
import sympy as sym
import numpy as np
from IPython.display import display, Math
```

```python
A = np.array([[1,2],[3,4]])

print(A)
display(Math(sym.latex(sym.sympify(A))))
```

```python
mat = np.zeros([4,6])
mat
```

```python
mat[0,1] = 2
mat
```

```python
mat[2,4] = 7
mat
```

```python
x = sym.symbols('x')
y = sym.symbols('y')

fm = sym.matrices.Matrix([[2,3],[4,5]])
fm2 = sym.matrices.Matrix([x, y])
display(Math(sym.latex(sym.sympify(fm))))
display(Math(sym.latex(sym.sympify(fm2))))
fm3 = fm * fm2
display(Math(sym.latex(sym.sympify(fm3))))
```

```python
fm = np.array([[2,3],[4,1]])
fm2 = np.array([4,8])

display(Math(sym.latex(sym.sympify(fm))))
display(Math(sym.latex(sym.sympify(fm2))))

fm3 = fm * fm2
display(Math(sym.latex(sym.sympify(fm3))))
```

```python
numrange = range(0,2)

for rowi in numrange:
    for coli in numrange:
        mat[rowi,coli] = (-1)**(rowi+coli)
mat
```

```python
x, y = sym.symbols('x y')
Fxy = (4+x)*(2-y)
display(Math(sym.latex(Fxy)))

xyset = range(0,3)

outmat = np.zeros([len(xyset), len(xyset)])

for i in xyset:
    for j in xyset:
        outmat[i,j] = Fxy.subs({x:i, y:j})

display(Math(sym.latex(sym.sympify(outmat))))
```
