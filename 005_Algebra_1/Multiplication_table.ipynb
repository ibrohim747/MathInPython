```python
import sympy as sym
import numpy as np
from IPython.display import display, Math
```

My version

```python
xyset = range(0,11)

outmat = np.zeros([len(xyset), len(xyset)])

for i in xyset:
    for j in xyset:
        outmat[i,j] = i * j

outmat2 = np.zeros([10,10])

for i in range(0,10):
    for j in range(0,10):
        outmat2[i,j] = outmat[i+1,j+1]

display(Math(sym.latex(sym.sympify(outmat2))))
```

Other version

```python
outmat3 = np.zeros([len(xyset), len(xyset)])

for i in xyset:
    for j in xyset:
        outmat3[i-1,j-1] = i * j

display(Math(sym.latex(sym.sympify(outmat3))))
```
