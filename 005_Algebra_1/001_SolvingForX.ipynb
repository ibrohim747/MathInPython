```python
import sympy as sym
from IPython.display import display, Math
```

```python
x = sym.symbols('x')
expr = 2*x + 4 - 9

sym.solve(expr)
```

```python
display(Math('\\text{The solution is } %s \\text{ is x= } %g' %(sym.latex(expr), sym.solve(expr)[0])))
```

```python
display(Math('3q + \\frac{4}{q} + 3 = 5q + \\frac{1}{q} + 1'))
q = sym.symbols('q')
eq = 3*q + 4/q + 3 - (5*q + 1/q + 1)
b = sym.solve(eq, q)
display(Math(sym.latex(eq.simplify()))) #Simplyfy
display(Math('q = ' + sym.latex(b)))
```

```python
display(Math('2q + 3q^2 - \\frac{5}{q} - \\frac{4}{q^3}'))
q = sym.symbols('q')
eq = 2*q + 3*q**2 - 5/q - 4/(q**3)
print("")
display(Math(sym.latex(eq)))
print("")
display(Math(sym.latex(sym.simplify(eq)))) #Simplyfy
print("")
display(Math(sym.latex(sym.cancel(eq)))) #Cancel
print("")
sym.solve(eq, q)
```

```python
display(Math('\\frac{\sqrt{3} + \sqrt{15}q}{\sqrt{2} + \sqrt{10}q}'))
q = sym.symbols("q")
eq = (sym.sqrt(3) + sym.sqrt(15)*q) / (sym.sqrt(2) + sym.sqrt(10)*q)
print("")
display(Math(sym.latex(eq)))
print("")
display(Math(sym.latex(sym.simplify(eq))))   #q was drop out
print("")
sym.solve(eq, q)
```
