```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#y = m*x + b

x = [-5,5]
m = 2
b = 1

y = [340, 2345]

for i in range(0, len(x)):
    y[i] = m*x[i] + b

plt.plot(x,y, label='y=%sx+%s' %(m,b))
plt.axis('square')
plt.grid(True)
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(), [0,0], 'k--')
plt.plot([0,0], axis.get_ylim(), 'k--')

plt.legend()

plt.show()
```

```python
#y = m*x + b

x = [-5,5]
m = 0.7
b = -2
for i in range(0, len(x)):
    y[i] = m*x[i] + b


x2 = [-5,5]
m2 = -1.25
b2 = 0.75
y2 = [10, 10]
for i2 in range(0, len(x2)):
    y2[i2] = m2*x2[i2] + b2

plt.plot(x,y, 'b', label='y=%sx+%s' %(m,b))
plt.axis('square')
plt.grid(True)
plt.xlim(x)
plt.ylim(x)

plt.plot(x2,y2, 'y', label='y=%sx+%s' %(m2,b2))
plt.axis('square')
plt.grid(True)
plt.xlim(x2)
plt.ylim(x2)

axis = plt.gca()
plt.plot(axis.get_xlim(), [0,0], 'k--')
plt.plot([0,0], axis.get_ylim(), 'k--')

plt.legend()

plt.show()
```
