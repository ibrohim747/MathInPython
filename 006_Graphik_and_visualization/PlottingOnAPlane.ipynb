```python
import matplotlib.pyplot as plt
```

```python
x = 3
y = 5

plt.plot(x, y, 'gs')
plt.axis('square')
plt.axis([-6, 6, -6, 6])
plt.grid()

plt.show()
```

```python
x = [-4, 5, 8, 3, 7, 1, 9]
y = [0, 5, 7, 0, 1, 2, 6]

for i in range(0, len(x)):
    plt.plot(x[i], y[i], 'gs', label=f'({x[i]}, {y[i]})')




plt.axis('square')
plt.legend()
plt.axis([-10, 10, -10, 10])
plt.grid()
plt.show()
```

```python
import sympy as sym
x = sym.Symbol('x')
numrange = range(-10, 11)

for x in numrange:
    y = x**2 - 3*x
    plt.plot(x, y, 'rs', label=f'({x}, {y})')

plt.xlabel('X')
plt.ylabel('f(x) = x**2 - 3*x')
plt.show()
```
