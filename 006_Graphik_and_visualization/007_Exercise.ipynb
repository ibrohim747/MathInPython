```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10
A = np.zeros((N, N))

for i in range(0, N):    # np.shape(A) == (10,14)
    for j in range(0, N):
        A [i, j] = (-1)**(i+j)


print(A)
plt.set_cmap('gray')
plt.tick_params(labelleft=False, labelbottom=False)
plt.imshow(A)
```
