```python
import sympy as sym
import sympy.plotting.plot as symplot
```

```python
from sympy.abc import x

y = x**2

p = symplot(y, show = "False")
p.xlim = [0,50]
p.ylim = [0,50]
p[0].line_color ='m'
p.title = "Nice"

p.show()
```

```python
x, a = sym.symbols('x a')

expr = a/x

p = symplot(expr.subs(a,1), (x, -5, 5), show=False)

p.ylim= [-5, 5]
p.show()
```

```python
from sympy.abc import x, y, a

y = a / (x**2 - a)

p = symplot(y.subs(a,1), (x,-5,5), show=False)
p[0].label = '$%s$' %(sym.latex(y.subs(a,1)))

for i in range(2,5):
    p.extend( symplot(y.subs(a,i), show=False))
    p[i-1].label = '$%s$' %(sym.latex(y.subs(a,i)))

p.ylim= [-6, 6]
p.legend = True
p.show()
```
