```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = range(-3,4)
y = np.zeros(len(x))

for i in range(0, len(x)):
    y[i] = 2 - x[i]**2

plt.plot(x,y, 's-')
plt.show()
```

```python
#y = 2 + np.sqrt(abs(x))

for i in range(0, len(x)):
    y[i] = 2 + np.sqrt(abs(x[i]))

plt.plot(x,y, 's-')
plt.show()
```
