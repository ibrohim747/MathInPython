```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = [[1,2],[1,4]]

plt.xticks([0,1])
plt.yticks([0,1])
plt.imshow(A)    #Image show
```

```python
A = np.zeros((10, 14))

for i in range(0, np.shape(A)[0]):    # np.shape(A) == (10,14)
    for j in range(0, np.shape(A)[1]):
        A [i, j] = 3*i - 4*j


for i in range(0, np.shape(A)[0]):
    for j in range(0, np.shape(A)[1]):
        plt.text(j, i, int(A[i, j]), horizontalalignment='center', verticalalignment='center')



print(A)
plt.set_cmap('gray')    #'twilight_shifted'

plt.savefig('test.pdf')

plt.imshow(A)
```
